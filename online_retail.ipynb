<b>Detecting outliers</b>

```python
sns.boxplot(x=df_online_retail['UnitPrice'])
```

```python
price_column = df_online_retail['UnitPrice']

# Calculate quartiles for the 'UnitPrice' column
Q1_price = price_column.quantile(0.25)
Q3_price = price_column.quantile(0.75)
IQR_price = Q3_price - Q1_price
print(IQR_price)
```

```python
# Create a mask to filter rows based on IQR
mask_price = (price_column < (Q1_price - 1.5 * IQR_price)) | (price_column > (Q3_price + 1.5 * IQR_price))

# Apply the mask to filter out rows
df_online_retail = df_online_retail[~mask_price]
df_online_retail.shape
```

```python
sns.boxplot(x=df_online_retail['UnitPrice'])
```

```python
# Assuming df_online_retail is your DataFrame and 'InvoiceDate' is the date column
# Convert 'InvoiceDate' to datetime if it's not already in datetime format
df_online_retail['InvoiceDate'] = pd.to_datetime(df_online_retail['InvoiceDate'])

# Extract month from 'InvoiceDate'
df_online_retail['Month'] = df_online_retail['InvoiceDate'].dt.month

# Count transactions per month
transactions_per_month = df_online_retail['Month'].value_counts().sort_index()

# Find the busiest month (month with the highest number of transactions)
busiest_month = transactions_per_month.idxmax()

# Plotting the data
plt.figure(figsize=(10, 6))
transactions_per_month.plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions per Month')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.axvline(x=busiest_month - 1, color='red', linestyle='--', linewidth=1.5, label='Busiest Month')
plt.legend()
plt.tight_layout()
plt.show()

# Print the busiest month
print(f"The busiest month is: {busiest_month}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_online_retail is your DataFrame and 'InvoiceDate' is the date column
# Convert 'InvoiceDate' to datetime if it's not already in datetime format
df_online_retail['InvoiceDate'] = pd.to_datetime(df_online_retail['InvoiceDate'])

# Extract day of the week from 'InvoiceDate' (Monday=0, Sunday=6)
df_online_retail['DayOfWeek'] = df_online_retail['InvoiceDate'].dt.dayofweek

# Map day of week to actual day names for better readability
day_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df_online_retail['DayOfWeek'] = df_online_retail['DayOfWeek'].map(day_map)

# Count transactions per day of the week
transactions_per_day = df_online_retail['DayOfWeek'].value_counts().reindex(day_map.values(), fill_value=0)

# Find the busiest day of the week (day with the highest number of transactions)
busiest_day = transactions_per_day.idxmax()

# Print the busiest day of the week
print(f"The busiest day of the week is: {busiest_day}")

# Plotting the data
plt.figure(figsize=(10, 6))
transactions_per_day.plot(kind='bar', color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions per Day of the Week')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```
